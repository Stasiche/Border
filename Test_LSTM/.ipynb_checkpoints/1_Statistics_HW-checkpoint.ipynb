# Задание 2

## Стратификация

Начнем со следующего примера:

Предположим, что нам нужно оценить среднее число голосов, поданных за каждого кандидата на выборах. Предположим, что в стране 3 города: в городе А живёт 1 миллион заводских рабочих, в городе Б живёт 2 миллиона офисных работников, а в городе В живёт 3 миллиона пенсионеров. Мы можем выбрать получение случайной выборки размером в 60 голосов со всей популяции, но есть некоторая вероятность, что случайная выборка окажется плохо сбалансирована по этим городам и, следовательно, будет необъективна и малополезна («средняя температура по больнице»), вызывая значительную погрешность в оценке. Взамен этого, если мы выберем использовать простую случайную выборку в 10, 20 и 30 голосов из городов А, Б и В соответственно, мы можем получить меньшую погрешность в оценке при том же общем размере выборки. Такой прием называется стратификацией.

Подробнее о стратификации [здесь](https://compscicenter.ru/media/slides/math_stat_2017_spring/2017_02_27_math_stat_2017_spring_qWLJs8i.pdf)

### Задача

Предположим, что генеральная совокупность - смесь 3х нормально распределенных СВ. Иными словами, генеральную совокупность можно разбить на 3 страты.
$$
    F(X) = a_1 F(X_1) + a_2 F(X_2) + a_3 F(X_3)
$$

** Требуется:**
* Вывести формулы и подсчитать точечные оценки математического ожидания и дисперсии для среднего значения подвыборок, образованных разными способами:
    * случайный выбор из всей генеральной совокупности
    * случайный выбор из страт пропорционально долям
    * случайный выбор согласно оптимальному разбиению на подсэмплы
Оценки вычисляются из результатов повторения эксперимента 1000 раз

Эксперименты провести для 3ех разных случаев:
* все страты имеют одинаковые мат. ожидания и дисперсии
* страты имеют разные мат.ожидания, но одинаковые дисперсии
* страты имеют разные и мат. ожидания и дисперсии.

Также провести эксперимент для разных размеров генеральной совокупности: 40 и 500. Параметры нормальных распределений, доли и размер подвыборки задайте самостоятельно

Внутри k-ой страты выделим подвыборку объема $n_k$ (нумерация между стратами несквозная)

По определению, выборочное среднее определяется следующим соотношением:
\begin{equation}
    \overline{X}_k \stackrel{\mathrm{def}}{=} \frac{1}{n_k} \sum\limits_{i=1}^{n_k} X_i \\
\end{equation}
Найдем математическое ожидание выборочного среднего (математическое ожидание элементов страты номер k равно $\mu_k$). Учитываем линейности математического ожидания: 
\begin{equation}
    E\overline{X}_k = \frac{1}{n_k}\sum\limits_{i=1}^{n_k} EX_i = \frac{n_k*\mu_k}{n_k} = \mu_k
\end{equation}
Найдем дисперсию выборочного среднего ($\sigma_k$ - дисперсия элементов страты номер k):
\begin{equation}
    D\overline{X}_k = D \left[ \frac{1}{n_k}\sum\limits_{i=1}^{n_k} X_i \right] = \left[\text{свойство дисперсии}\right] = \frac{1}{{n_k}^2}D \left[ \sum\limits_{i=1}^{n_k} X_i \right] = \left[\text{так как $X_i$ -  независимые СВ}\right] = \frac{1}{{n_k}^2}\sum\limits_{i=1}^{n_k} \left[ DX_i \right] = \frac{\sigma_k}{n_k}
\end{equation}
Введем следующую оценку выборочного среднего (L - суммарное количество страт, $W_l$ - доля страты в генеральной совокупности):
\begin{equation}
    \overline{X}_s = \sum\limits_{l=1}^{L} W_l \overline{X}_l
\end{equation}
Если объемы сэмплов выбраны оптимально, то есть $n_k = n \frac{W_k\sigma_k}{\sum\limits_{l=1}^{L} W_l\sigma_l}$, где $n = \sum\limits_{k=1}^{L} n_k $-  суммарное количество выбранных элементов, то будем называть ту же по формулировке оценку $\overline{X}_{SO}$.

Если объемы сэмплов выбраны пропорционально, то есть $n_k = nW_k\sigma_k$, где $n = \sum\limits_{k=1}^{L} n_k $-  суммарное количество выбранных элементов, то будем называть ту же по формулировке оценку $\overline{X}_{SP}$.
Так как математическое ожидание линейно, то математическое ожидание точечной оценки:
\begin{equation}
     E\overline{X}_S = \sum\limits_{i=1}^{L} W_iE\overline{X}_i = \sum\limits_{i=1}^{L} W_i\mu_i
\end{equation}
Используя свойство дисперсии линейной комбинации случайных величин и тот факт, что ввиду независимости $\overline{X}_k$ сумма всевозможных ковариаций равна нулю можно показать, чему будет равна дисперсия точечной оценки
\begin{equation}
     D\overline{X}_S = \sum\limits_{i=1}^{L} {W_i}^2D\overline{X}_i = \sum\limits_{i=1}^{L}\left[ {W_i}^2 \left(\frac{1}{n_i}\right)\left(1 - \frac{n_i - 1}{N_i - 1}\right){\sigma_i}^2\right]
\end{equation}
Пусть $n_i  \ll  N_i$, тогда:
\begin{equation}
     D\overline{X}_S  \sim \sum\limits_{i=1}^{L}\left[\frac{{W_i}^2{\sigma_i}^2}{n_i}\right]
\end{equation}
Тогда простой подстановкой получаются следующие дисперсии оценок:
\begin{equation}
     D\overline{X}_{SP} = \frac{1}{n} \sum\limits_{i=1}^{L}\left[W_i{\sigma_i}^2\right] 
\end{equation}
\begin{equation}
     D\overline{X}_{SO} = \frac{1}{n}\left( \sum\limits_{i=1}^{L}\left[W_i\sigma_i\right] \right)^2
\end{equation}

````python
class Experiment:

    def __init__(self, means, stds, random_state=1):
        """Инициализирует наш эксперимент и сохраняем заданные распределения
        
        :param means: Список матожиданий для нормальных распределний
        :param stds: Список стандартных отклонений для нормальных распределний
        :param random_state: Параметр фиксирующий случайность. Нужен для того, чтобы при воспроизведении
        эксперимента повторно с теми же входными параметрами, результаты оставались такими же
        """
        self.strats = [st.norm(mean, std) for mean, std in zip(means, stds)]
        
    def sample(self, sizes):
        """Создает выборку генеральной совокупности
        
        :param sizes: Список с размерами выборок соответсвующих им нормальных распределений
        """
        self.strats_samples = [rv.rvs(size) for rv, size in zip(self.strats, sizes)]
        self.general_samples = np.hstack(self.strats_samples)
        self.N = self.general_samples.shape[0]
        self.w = [ss.shape[0]/self.N for ss in self.strats_samples]
        # print(self.general_samples)
        
    def random_subsampling(self, size):
        """Создает случайную подвыборку из генеральной совокупности
        
        :param sizes: размер подвыборки
        """
        ran_subsamp = self.general_samples[np.random.randint(self.N, size=size)]
        np.random.shuffle(ran_subsamp)
        # print(ran_subsamp)
        return ran_subsamp
        
    def proportional_subsampling(self, size):
        """Создает подвыборку с кол-вом элементов, пропроциональным долям страт
        
        :param sizes: размер подвыборки
        """
        prop_subsamp = []
        for i, ss in enumerate(self.strats_samples):
            size_ss = ss.shape[0]
            n_k = (int)(round(self.w[i]*size))
            prop_subsamp = np.append(prop_subsamp,ss[np.random.randint(size_ss, size=n_k)])
        np.random.shuffle(prop_subsamp)
        # print(prop_subsamp)
        return prop_subsamp
    
    def optimal_subsampling(self, size):
        """Создает подвыборку с оптимальным кол-вом элементов относительно страт
        
        :param sizes: размер подвыборки
        """
        opt_subsamp = []
        tmp_coef = np.array([self.w[i]*self.strats[i].std() for i, ss in enumerate(self.strats_samples)])
        tmp_coef = tmp_coef.sum()
        for i, ss in enumerate(self.strats_samples):
            size_ss = ss.shape[0]
            n_k = (int)(round(size*self.w[i]*self.strats[i].std()/tmp_coef))
            opt_subsamp = np.append(opt_subsamp,ss[np.random.randint(size_ss, size=n_k)])
        np.random.shuffle(opt_subsamp)
        # print(opt_subsamp)
        return opt_subsamp
        
    def run_experiments(self, subsampling_method, size_subsampling, n_experiments=1000):
        """Проводит серию экпериментов и сохраняет результаты
        
        :param n_experiments: кол-во запусков эксперимента
        """
        results = {'exs': [], 'dxs': []}
        dict_method = {'rs': self.random_subsampling, 'ps': self.proportional_subsampling, 'os':self.optimal_subsampling}
        sm = dict_method.get(subsampling_method)
        for step in range(n_experiments):
            if sm is not None:
                sub_samp = sm(size_subsampling) 
            else:
                print('Bad sampling method')
                return -1
            
            exs = np.array([self.w[i]*self.strats[i].std() for i, ss in enumerate(sub_samp)]).sum()
            
            dxs = None
            if subsampling_method == 'rs':
                dxs = np.array([np.power(self.w[i]*self.strats[i].std(),2)/ss.shape[0] for i, ss in enumerate(sub_samp)]).sum()
            elif subsampling_method == 'ps':
                dxs = np.array([self.w[i]*np.power(self.strats[i].std(),2) for i, ss in enumerate(sub_samp)]).sum()/size_subsampling
            elif subsampling_method == 'os':
                dxs = np.power(np.array([self.w[i]*self.strats[i].std() for i, ss in enumerate(sub_samp)]).sum(),2)/size_subsampling
            
            if dxs is not None:
                results['exs'].append(exs)
                results['dxs'].append(dxs)
                
        self.results = results
````

# Задание 3

## Последовательный критерий Вальда (критерий [SPRT](https://en.wikipedia.org/wiki/Sequential_probability_ratio_test))

[Видео с объяснением](https://www.youtube.com/watch?v=4NYbTUIkBbU&index=4&list=PLlb7e2G7aSpRG_Ve6kuxc9Biuwyf42kJx) (начало 1:06:20) 

### Задача 
* Создайте синтетические данные полученные путем смешивания 2 выборок из нормально распределеных СВ с разными матожиданиями, но одинаковыми дисперсиями
* Примените к сгенерированным данным последовательный критерий Вальда силы $(\alpha, \beta)$ (критерий Вальда должен быть реализован самостоятельно). Для этого:
    * Задайте нулевую и альтернативную гипотезы о равенствах матожиданий конкретным величинам
    * Задайте $\alpha, \beta$
    * Выпишите формулы аппроксимации границы
    * Провести эксперимент
    * Результат эксперимента отобразить на графике
* Проанализируйте поведение критерия при разных долях компонент в смеси
* Какой объем выборок потребовался в каждом случае?
* Выпишите нижнюю оценку для среднего числа наблюдений при заданных $(\alpha, \beta)$ в случае справедливости нулевой и альтернативной гипотез

````python
class Experiment_task_3:

    def __init__(self, means, stds, n1, n2, m_hip_0, m_hip_1, alpha, beta):
        sizes = [n1, n2]
        self.std_total = math.sqrt(stds[0] ** 2 + stds[1] ** 2)

        self.m_hip_0 = m_hip_0
        self.m_hip_1 = m_hip_1
        # генерируем примеры, откуда в последствии будем брать X
        self.strats = [st.norm(mean, std) for mean, std in zip(means, stds)]
        self.samples = np.hstack([rv.rvs(size) for rv, size in zip(self.strats, sizes)].copy())
        np.random.shuffle(self.samples)

        # апроксимируем границы
        self.c0 = beta / (1 - alpha)
        self.c1 = (1 - beta) / alpha

    # функция вычисления функции правдоподобия 
    def lh(self, X_, m_hip):
        # функция плотности вероятности результирующего распределения
        g = st.norm(m_hip, self.std_total)
        return np.prod(g.pdf(X_))
    
    def run_experiment(self):
        # инициализируем значения для проведения эксперимента
        X = []  # набор "проведенных" замеров
        z_total = []  # значения статистики
        # проводим эксперимент
        for el in self.samples:
            X.append(el)
            z = self.lh(X, self.m_hip_0) / self.lh(X, self.m_hip_1)

            if not(self.c0 <= z <= self.c1):
                print('Принимаем %s: математическое ожидание распределения равно %f' % 
                (('H0', self.m_hip_0) if self.c0 > z else ('H1', self.m_hip_1)))
                break
                      
            z_total.append(z)
            # print(z)
        print('Used %d elements.' % len(z_total))
        plt.plot(z_total)
        plt.show()
````

````python
std_0 = 5.9
m_hip_0_0 = 0.2
m_hip_1_0 = -0.3


print('Experiment #1')
# propotion 1:1
exp1 = Experiment_task_3(means=(-10, 10), stds=(std_0, std_0), n1=500, n2=500, m_hip_0=m_hip_0_0, m_hip_1=m_hip_1_0, alpha=1e-4, beta=1e-3)
exp1.run_experiment()

print('Experiment #2')
# propotion 1:3
exp2 = Experiment_task_3(means=(-10, 10), stds=(std_0, std_0), n1=250, n2=750, m_hip_0=m_hip_0_0, m_hip_1=m_hip_1_0, alpha=1e-4, beta=1e-3)
exp2.run_experiment()

print('Experiment #3')
# propotion 1:9
exp3 = Experiment_task_3(means=(-10, 10), stds=(std_0, std_0), n1=100, n2=900, m_hip_0=m_hip_0_0, m_hip_1=m_hip_1_0, alpha=1e-4, beta=1e-3)
exp3.run_experiment()

print('Experiment #4')
# propotion 3:1
exp4 = Experiment_task_3(means=(-10, 10), stds=(std_0, std_0), n1=750, n2=250, m_hip_0=m_hip_0_0, m_hip_1=m_hip_1_0, alpha=1e-4, beta=1e-3)
exp4.run_experiment()

print('Experiment #5')
# propotion 9:1
exp5 = Experiment_task_3(means=(-10, 10), stds=(std_0, std_0), n1=900, n2=100, m_hip_0=m_hip_0_0, m_hip_1=m_hip_1_0, alpha=1e-4, beta=1e-3)
exp5.run_experiment()
````

Эксперементальная база не так велика, но вроде как можно делать вывод, что чем больше отношение долей страт  в общей выборке, тем меньше нужно примеров для того, чтобы критерий сработал.

Оценка среднего числа наблюдений при заданных $(\alpha,\beta)$  при справедливости нулевой гипотезы:
\begin{equation}
    {E_0}_\nu(\alpha,\beta) \geq - \frac{\omega(\alpha,\beta)}{{E_H}_{0}}
\end{equation}
Альтернативной гипотезы:
\begin{equation}
    {E_1}_\nu(\alpha,\beta) \geq \frac{\omega(\beta,\alpha)}{{E_H}_{1}}
\end{equation}
Чтобы вычислить оценку среднего числа наблюдений, вычислим промежуточные значения:
\begin{equation}
    {E_h}_j = \frac{1}{ \sigma \sqrt{2 \pi } \sigma}\exp \left(-\frac{(x-m_j)^2}{2 \sigma^2}\right)*\left(-\frac{(m_0-m_1) (-m0-m1+2 x)}{2 \sigma^2}\right).
\end{equation}
\begin{equation}
    {E_h}_0 = -0.00359092
\end{equation}
\begin{equation}
    {E_h}_1 = 0.00359092
\end{equation}
\begin{equation}
    \omega(\alpha, \beta) = \omega(10^{-4}, 10^{-3}) = 6.90604
\end{equation}
\begin{equation}
    \omega(\beta, \alpha) =\omega(10^{-3}, 10^{-4}) = 9.19322
\end{equation}

Таким образом,  оценка среднего числа наблюдений в случае справедливости нулевой гипотезы равна:

### Задача 2
Используя критерий [хи-квадрат](http://www.machinelearning.ru/wiki/index.php?title=%D0%A2%D0%B0%D0%B1%D0%BB%D0%B8%D1%86%D0%B0_%D1%81%D0%BE%D0%BF%D1%80%D1%8F%D0%B6%D0%B5%D0%BD%D0%BD%D0%BE%D1%81%D1%82%D0%B8), проверить, есть ли различия в выборе авто между мужчинами и женщинами (столбцы `Sex` и `PreferCar`)

** Данные в файле `HW1_task4_car_prefs_data.tsv`**

````python
df = pd.read_csv('./data/HW1_task4_brain_data.tsv', sep='\t')
df_male = df[df['Gender'] == 'Male']
df_female = df[df['Gender'] == 'Female']

X_total = df['FSIQ']
X_male = df_male['FSIQ']
X_female = df_female['FSIQ']

Y_total = df['MRI_Count']
Y_male = df_male['MRI_Count']
Y_female = df_female['MRI_Count']

pairs = ((X_total,Y_total,'Total'), (X_male, Y_male,'Male'), (X_female, Y_female,'Female'))
for X,Y,sex in pairs:
    n = X.shape[0]
    X_mean = X.mean()
    Y_mean = Y.mean()

    X_std = X.std()
    Y_std = Y.std()

    tmp = ((X - X_mean)*(Y - Y_mean)).sum()
    
    r = tmp/(X_std*Y_std*n)
    
    t = r/math.sqrt(1-r**2)*math.sqrt(n-2)
    z = 0.5*math.log((1+r)/(1-r))
    
    tmp = 1.96/math.sqrt(n-3)
    ci_z = (z - tmp, z + tmp)
    hip = -1
    if (z > ci_z[0]) and (z < ci_z[1]):
        hip = 'H0'
    else:
        hip = 'H1'
    print('%s \n r: %f, t: %f, z: %f, hip: %s' % (sex, r,t,z,hip))
````

````python
df = pd.read_csv('./data/HW1_task4_car_prefs_data.tsv', sep='\t')
ct = pd.crosstab(df['Sex'],df['PreferCar'])

o_i = ct.sum(axis=0)
n_j = ct.sum(axis=1)
# H0 - мужчины и женщины выбирают машины "одинаково" (то есть с похожими распредлениями)
t = 1/np.prod(n_j)*(np.power((n_j.loc[1]*ct.loc[1] - n_j.loc[2]*ct.loc[2]),2)/o_i).sum()
print(t)
````

В соответствии с условиями применения двустороннего критерия хи-квадрат для одной степени свободы и уровня значимости a=0.05 найдем $Т_{критич}$ = 7.815. То есть отвергаем нулевую гипотезу и делаем вывод, что женщины и мужчины по-разному выбирают автомобили.
